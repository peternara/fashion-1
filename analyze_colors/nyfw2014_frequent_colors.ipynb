### 1. OpenCV Haar-Cascade implementation

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pyl
import cv2
```

```python
'''Get the already trained classifier from an XML file
   Input:  path to the XML file
   Output: haar-cascade classifier'''
def trained_classifier(file):
    return cv2.CascadeClassifier(file)

'''Run the classifier against the provided image
   Input:  image, trained classifier
   Output: numpy n-dimensional array with the detected faces
           every face is defined by the array (x,y,width,height)'''
def face_detector(image, classifier):
    return classifier.detectMultiScale(image, scaleFactor=1.3, minNeighbors=5)

'''Draw a rectangle on the image'''
def draw_rects(img, rects):
    for x, y, w, h in rects:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)
```

```python
#load the image
def load_image(filepath):
    return pyl.imread(filepath)

#transform to gray scale
def gray_image(image):
    return cv2.cvtColor(img, cv2.cv.CV_BGR2GRAY) #openCV read the color values as BGR instead of RGB that uses matplotlib
    
#load the classifier
cascade_classifier =  trained_classifier('opencv-2.4.8/data/haarcascades/haarcascade_frontalface_alt.xml')
```

```python
#detect the faces on image
img = load_image('fall2014rtw_ny/carolina/109.jpg')
gray = gray_image(img)
faces = face_detector(gray, cascade_classifier)
draw_rects(img, faces)
```

### 2. Crop the image selecting the ROI

```python
'''Calculate the area of a face. Used as sorted key function on select_face method
   Input: numpy array with a face
   Output: area'''
def area(face):
    return face[2] * face[3]

'''Select the bigger face on the image
   Input: numpy n-dim array with the faces
   Output: numpy array with the selected face'''
def select_face(faces):
    return sorted(faces, key=area, reverse=True)[0]
```

```python
'''Crop the image by selecting the ROI
   Input: image, array of faces
   Output: ROI cropped image'''
def crop_image(image, faces):
    yt, xt, rgb = image.shape
    x,y,w,h = select_face(faces)
    return image[y+h:yt-200,x-150:x+w+150]
```

```python
img_cropped = crop_image(img, faces)
plt.figure(1)
plt.subplot(121)
plt.imshow(img)
plt.subplot(122)
plt.imshow(img_cropped)
plt.show()
#pyl.imsave('fall2014rtw_ny/carolina/109_faces.jpg', img)
```

### 3. Apply a kmeans classifier to select the most frequent color on the image

```python
import scipy as sp
from scipy.cluster.vq import kmeans,vq
```

```python
'''Apply kmeans algorithm.
   Input:  image, number of clusters
   Output: centroids, vector with the index of the centroid assigned to each pixel, clusters'''
def perform_kmeans(image, k):
    rows, cols, rgb = image.shape
    # reshape the image in a single row array of RGB pixels
    image_row = np.reshape(image,(rows * cols, 3))
    # perform the clustering
    centroids,_ = kmeans(image_row, k)
    # vector quantization, assign to each pixel the index of the nearest centroid (k=1..4)
    qnt,_ = vq(image_row,centroids)
    # reshape the qnt vector to the original image shape
    image_centers_id = np.reshape(qnt,(rows, cols))
    # assign the centroid value to each pixel
    clustered = centroids[image_centers_id]
    return centroids, qnt, clustered
```

```python
centroids, qnt, clustered = perform_kmeans(img_cropped, 4)
```

```python
'''Count the number of pixels associated to each centroid
   Input: quantization vector, centroids
   Output: k length array with the count of pixels per centroid''' 
def count_colors(qnt, centroids):
    # count number of pixels on each cluster
    counts, bins = sp.histogram(qnt, len(centroids))
    return counts, bins

'''Transform a vector of RGB colors to its hex value
   Input: array of RGB values
   Outpur: array of hex values'''
def colors_to_hex(colors):
    return [''.join(map(chr, color)).encode('hex') for color in colors]
```

```python
counts, bins = count_colors(qnt, centroids)
colors = colors_to_hex(centroids)

# visualizing the number of pixels associated with each centroid
fig = plt.figure(3, figsize=(8,5))
ax = plt.subplot(111)
width = 0.8
barlist = ax.bar(range(len(counts)), counts)
for i in range(0,4):
    barlist[i].set_color(centroids[i]/255.)
ax.set_xticks(np.arange(len(counts)) + width/2)
ax.set_xticklabels(colors, rotation=90)
ax.set_title("       #Pixels associated with each centroid")
plt.ylabel("#Pixels")
plt.xlabel("Centroid colors (hex value)")
plt.show()
#fig.subplots_adjust(bottom=0.2)
#fig.savefig('figure1.png', dpi = (200))
```

```python
'''Select the most repetead color
   Input: centroids, vector with the number of occurrences for each color
   Output: most repetead centroid'''
def predominant_color(centroids, counts):
    index_max = sp.argmax(counts)
    return centroids[index_max]
```

### 4. Repeat the process for all the images in the collection

```python
files = !ls fall2014rtw_ny/carolina
```

```python
selected_color = np.zeros((len(files),3))
```

```python
for i, file in enumerate(files):
  # Step0. Load the image and transform to gray scale
    img = load_image('fall2014rtw_ny/carolina/' + file)
    gray = gray_image(img)
    
  # Step1. Apply the haar-cascade classifier for detect the faces
    faces = face_detector(gray, cascade_classifier)
    
  # Step2. Crop the ROI
    img_cropped = crop_image(img, faces)
    
  # Step3. Apply the kmeans classifier (k = 4) for clustering the colors and select the most frequent
    centroids, qnt, clustered = perform_kmeans(img_cropped, 4)
    # count number of pixels associated to each cluster
    counts, bins = count_colors(qnt, centroids)  
    # select the most repeated color
    selected_color[i] = predominant_color(centroids, counts)
```

```python
# visualizing the most repeated color of each image into the RGB space
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(4, figsize=(8,5))
ax = fig.gca(projection='3d')
ax.scatter(selected_color[:,0],selected_color[:,1],selected_color[:,2],c=selected_color/255.,s=75)

ax.set_title("       Selected color of each image in the RGB space")
ax.set_xlabel("Red value")
ax.set_ylabel("Green value")
ax.set_zlabel("Blue value")
#fig.subplots_adjust(bottom=0.2)
#fig.savefig('figure2.png', dpi = (200))
```

### 5. Select a smaller color space representation

```python
import webcolors
```

```python
'''Assign to an RGB color the closest CSS color by selecting the one with minimum Euclidian distance
   Input: RGB color
   Output: CSS color'''
def rgb_to_css(rgb_color):
    min_colors = {}
    for key, name in webcolors.css3_hex_to_names.items(): #key == hex value
        r_css, g_css, b_css = webcolors.name_to_rgb(name)
        rd = (r_css - rgb_color[0]) ** 2
        gd = (g_css - rgb_color[1]) ** 2
        bd = (b_css - rgb_color[2]) ** 2
        min_colors[(rd + gd + bd)] = name
    return min_colors[min(min_colors.keys())]

'''Get the name and the hex value of the RGB color in the CSS space. 
   If not direct match found, the closest color will be assigned
   Input: RGB color
   Outpur: CSS color'''
def get_color_name(rgb_color):
    try:
        color = webcolors.rgb_to_name(rgb_color)
    except ValueError:
        color = closest_color(rgb_color)
    hex_value = webcolors.name_to_hex(color)
    return color, hex_value
```

```python
css_colors = {}
for color in selected_color:
    c_name, hex_val = get_color_name(color)
    if c_name in css_colors:
        css_colors[c_name][0] += 1
    else:
        css_colors[c_name] = []
        css_colors[c_name].append(1)
        css_colors[c_name].append(hex_val)
```
